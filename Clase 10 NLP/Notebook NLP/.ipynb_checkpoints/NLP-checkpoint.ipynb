## Nube de Palabras

```python
from wordcloud import WordCloud
from nltk import word_tokenize
from collections import Counter
```

```python
def prepara_nube_palabras_menciones(df,cuenta,stop_words):
    df = df[df.screen_name == cuenta]
    
    df.drop_duplicates(inplace=True, subset=["tw_id"])
    
    lista = [word_tokenize(t) for t in df.full_text]
    flat_list = [item for sublist in lista for item in sublist]

    lower_tokens = [t.lower() for t in flat_list if t.isalpha()]

    stop_words = stop_words
    stop_words.append(cuenta[1:].lower())
    no_stop = [t for t in lower_tokens if t not in stop_words]

    text = " ".join(no_stop)
    return (text,no_stop)
    
def nube_palabras(text,background_color='white', figsize =(18,8) ):
    plt.figure(figsize=figsize)
    wordcloud = WordCloud(collocations=False,max_font_size=40,background_color='white',width=400,height=200).generate(text)

    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    
    plt.show()
    
```

```python
def prepara_texto(texto,stem = False):
    #Inicializo Stemmer
    stemmer = SnowballStemmer('spanish')
    #tokenizar, pasar a minuscula solo caracteres y realizar stemming
    texto = word_tokenize(texto)
    lower_tokens = [t.lower() for t in texto if t.isalpha()]
    if stem:
        no_stop = [stemmer.stem(t) for t in lower_tokens if t not in stop_words]
    else:
        no_stop = [t for t in lower_tokens if t not in stop_words]
    return no_stop

def preprocesamiento_texto(corpus,logging =False):
    #inicializa tamaño de corpus
    size_cor = len(corpus)
    avance = 0
    
    text_data = []
    for (texto,i) in zip(corpus,range(0,size_cor)):
        t = prepara_texto(texto)
        text_data.append(t)
        #logging
        if logging:
            if round(i*100/size_cor) != avance:
                print(round(i*100/size_cor),'%')
            avance = round(i*100/size_cor)
    return text_data

def intersection(lst1, lst2): 
    lst3 = [value for value in lst1 if value in lst2] 
    return lst3 

def remove_word_low_freq(text_data,n_min,n_max,logging = True):
    # Inicializar 
    size = len(text_data)
    avance = 0
    lista_aux = []
    # Pasa una lista de lista a una flat list
    flat_list = [item for sublist in text_data for item in sublist]
    # Crea un bag of words y se queda con los de baja frecuencia para luego removerlas
    counter = Counter(flat_list)
    baja_y_alta_freq = list({ key:value for key, value in counter.items() if value > n_min and value < n_max }.keys())
    for (l,i) in zip(text_data,range(0,len(text_data))):
        lista_aux.append(intersection(l,baja_y_alta_freq))
        #logging
        if logging:
            if round(i*100/size) != avance:
                print(round(i*100/size),'%')
            avance = round(i*100/size)
    return lista_aux

            
def n_gram(text_data,min_count = 5, threshold = 100):
    
    # Inicializar modelos bi y tri grams
    bigram = gensim.models.Phrases(text_data, min_count=min_count, threshold=threshold) # a thresholds más grandes menos frases.
    trigram = gensim.models.Phrases(bigram[text_data], threshold=threshold)  

    bigram_mod = gensim.models.phrases.Phraser(bigram)
    trigram_mod = gensim.models.phrases.Phraser(trigram)

    return trigram_mod[bigram_mod[text_data]]
```

```python
def crea_diccionario(text_data):
    #crea un diccionario
    dictionary = corpora.Dictionary(text_data)
    corpus = [dictionary.doc2bow(text) for text in text_data]
    return (dictionary,corpus)


dictionary,corpus = crea_diccionario(text_data)
```

```python
#modelo LDA
NUM_TOPICS = 10
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics = NUM_TOPICS, id2word=dictionary, passes=5)

```

```python
#desplega el resultado del modelo
topics = ldamodel.print_topics(num_words=14)
for topic in topics:
    print(topic,'\n')
```

```python
# Compute Coherence Score using c_v
coherence_model_lda = CoherenceModel(model=ldamodel, texts=text_data, dictionary=dictionary, coherence='c_v')
coherence_lda = coherence_model_lda.get_coherence()
print('\nCoherence Score: ', coherence_lda)

# Compute Coherence Score using UMass
coherence_model_lda = CoherenceModel(model=ldamodel, texts=text_data, dictionary=dictionary, coherence="u_mass")
coherence_lda = coherence_model_lda.get_coherence()
print('\nCoherence Score: ', coherence_lda)
```

```python
def compute_coherence_values(dictionary, corpus, texts, limit, start=2, step=3):

    coherence_values = []
    model_list = []
    for num_topics in range(start, limit, step):
        model=LdaModel(corpus=corpus, id2word=dictionary, num_topics=num_topics,random_state=42)
        model_list.append(model)
        coherencemodel = CoherenceModel(model=model, texts=texts, dictionary=dictionary, coherence='c_v')
        coherence_values.append(coherencemodel.get_coherence())
        print("Para n_topicos =",num_topics,"se obtiene un coherence value de: ",coherencemodel.get_coherence())

    return model_list, coherence_values
```

```python
from gensim.models.word2vec import Word2Vec
```

```python
# "window" es el tamaño de la ventana. windows = 10, usa 10 palabras a la izquierda y 10 palabras a la derecha
# "n_dim" es la dimension (i.e. el largo) de los vectores de word2vec
# "workers" es el numero de cores que usa en paralelo. Para aprovechar eso es necesario tener instalado Cython)
# "sample": word2vec filtra palabras que aparecen una fraccion mayor que "sample"
# "min_count": Word2vec filtra palabras con menos apariciones que  "min_count"
# "sg": para correr el Skipgram model (sg = 1), para correr el CBOW (sg = 0)
# para mas detalle ver: https://radimrehurek.com/gensim/models/word2vec.html
n_dim = 50
w2v_model = Word2Vec(text_data, workers=4,size=n_dim, min_count = 5, window = 5, sample = 1e-3,negative=10,sg=1)
```

```python
# Armo una matriz de distancias
distancias=np.zeros((len(palabras),len(palabras))) #matriz cuadrada
for i,ti in enumerate(palabras):
    for j,tj in enumerate(palabras):
        distancias[i,j] = abs(1-w2v_model.wv.similarity(ti,tj))
print (distancias.shape)
distancias
```

```python
# Reduccion de la dimensionalidad y visualizacion 
from sklearn.manifold import MDS
from sklearn.manifold import TSNE 
def visualize_embeddings(distancias,palabras,colores,perplexity):
    plt.figure(figsize=(35,8))
    # Reduccion de la dimensionalidad y visualizacion 
    mds = MDS(n_components=2, max_iter=3000, eps=1e-9, random_state=123,
                       dissimilarity="precomputed", n_jobs=4)
    Y = mds.fit(distancias).embedding_
    plt.subplot(1,2,1)
    plt.scatter(Y[:, 0], Y[:, 1],color="black",s=3)
    for label, x, y, color in zip(palabras, Y[:, 0], Y[:, 1],colores):
        plt.annotate(label, xy=(x, y), xytext=(0, 0),color=color, textcoords='offset points',size=13)
    plt.title("MDS")
```

```python
visualize_embeddings(distancias,palabras,colores,perplexity=10)
```
